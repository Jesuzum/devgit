Visualizacion con matplotlib y seaborn para data analyst

```python
# Importamos las librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creamos un DataFrame de ejemplo

df = pd.DataFrame({
    "edad": [23, 35, 30, 29, 40, 19, 50, 28],
    "salario": [12000, 18000, 16000, 14000, 25000, 8000, 30000, 15000],
    "ciudad": ["CDMX", "GDL", "CDMX", "MTY", "MTY", "GDL", "CDMX", "GDL"]
})
```

```python
# 1. Gráfico de líneas (con matplotlib)

plt.plot(df["edad"], df["salario"], marker='o')
plt.title("Salario según Edad")
plt.xlabel("Edad")
plt.ylabel("Salario")
plt.grid(True)
plt.show()
```

```python
# 2. Histograma (con matplotlib)

plt.hist(df["salario"], bins=5, color="skyblue", edgecolor="black")
plt.title("Distribución de salarios")
plt.xlabel("Salario")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# 3. Scatter plot (con matplotlib)

plt.scatter(df["edad"], df["salario"], color="red")
plt.title("Relación entre Edad y Salario")
plt.xlabel("Edad")
plt.ylabel("Salario")
plt.show()
```

```python
# 4. Countplot (con seaborn)

sns.countplot(x="ciudad", data=df)
plt.title("Cantidad de personas por ciudad")
plt.show()
```

```python
# 5. Boxplot (con seaborn)

sns.boxplot(x="ciudad", y="salario", data=df)
plt.title("Distribución de salario por ciudad")
plt.show()
```

```python
# 6. Heatmap (matriz de correlación con seaborn)

correlacion = df[["edad", "salario"]].corr()
sns.heatmap(correlacion, annot=True, cmap="coolwarm")
plt.title("Correlación entre variables")
plt.show()
```

Tipo de gráfico	    ¿Para qué sirve?
plot()	            Tendencias (líneas de tiempo, relaciones)
hist()	            Ver distribución de una variable numérica
scatter()	        Relación entre dos variables
countplot()	        Contar categorías (como un gráfico de barras)
boxplot()	        Ver valores mínimos, máximos, cuartiles y outliers
heatmap()	        Ver correlación entre variables numéricas
