```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

from sklearn.datasets import load_iris
```

```python
iris = load_iris(as_frame=True)
df = pd.DataFrame(data = iris.data.iloc[:,[2,3]], columns=iris.feature_names[2:4])
df['target'] = iris.target

print(df)

sns.pairplot(df, hue='target')
```

```python
i = df.index[df['target']==2].tolist()

df = df.drop(i)
sns.pairplot(df,hue="target")
```

```python
variableX = iris.data
variableX = iris.data.iloc[:, [2,3]]
print(f"Las variables independientes son \n: {variableX} \n" )

variableY = (iris.target)
print(f"Las variables dependientes son \n: {variableY} \n")
```

```python
variableX = variableX.drop(i)
variableY = variableY.drop(i)

print(f"El tamaño de las variables independientes es \n: {variableX.shape} \n")
print(f"El tamaño de las variables dependientes es \n: {variableY.shape} \n")
```

```python
x_train, x_test, y_train, y_test = train_test_split(variableX, variableY, test_size=0.2)

model = Perceptron(random_state=42)

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

f1 = f1_score(y_test, y_pred)
print(f"El f1 score del modelo es: {f1}")

percent = model.score(x_test, y_test)
print(f"El porcentaje de acierto del modelo es: {percent}")
```
